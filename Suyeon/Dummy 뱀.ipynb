뱀

Dummy라는 도스 게임 
- 뱀이 사과를 먹으면 몸 길이 늘어남
- 벽 또는 자기 자신이랑 부딪히면 game over
- NxN 정사각보드 
- 맨 처음은 맨위맨좌측에서 시작하고 길이는 1(한 칸), 뱀은 오른쪽을 보고있음 
- 뱀은 몸길이를 늘려 머리를 다음칸에 위치 시킴
- 이동한 칸에 사과가 있다면 사과는 없어지고 꼬리는 움직이지 않음 
- 사과가 없다면 몸길이를 줄여서 위치한 칸 비워줌(몸길이는 안 달라지만,, 머리는 움직였으니까)

입력조건
- 보드크기 N
- 사과의 수 K
- 사과의 위치 
- 뱀의 방향 변환 횟수 L
- 방향 변환 정보
 정수 X와 문자C로 구성,, 시작 X초 후 왼쪽(L) 또는 오른쪽(R)로 90도 방향 회전시킨다는 뜻 
 X는 10000이하 양의 정수

```python
#input 받기 

n = int(input())
k = int(input())

#맵정보
#일단 다 0으로 
data = [[0]*(n+1) for _ in range(n+1)]
info = [] #방향 회전 정보 
#사과가 있는 곳은 1로 표시 
for _ in range(k):
    a,b = map(int,input().split())
    data[a][b] = 1 

    
#방향 회전 정보 받기 
l = int(input())
for _ in range(l):
    x,c = input().split()
    info.append((int(x),c))
    
#방향설정
dx = [0,1,0,-1]
dy = [1,0,-1,0]

#방향 변환 함수 
def turn(direction,c):
    if c == "L":
        direction = (direction-1) % 4
    else:
        direction = (direction+1) % 4
    return direction 

#시뮬레이션 함수 
def sim():
    x,y=1,1 #뱀 초기 머리 위치 
    data[x][y] = 2 #뱀은 2로 표시 
    direction = 0 #동쪽 
    time = 0
    index = 0 #다음 회전 정보 
    q = [(x,y)] #뱀이 차지하고 있는 위치 정보 
    while True:
        nx = x + dx[direction]
        ny = y + dy[direction]
#맵 범위 안에 있고 뱀이 없는 위치라면
        if 1 <= nx and nx <= n and 1 <= ny and ny <= n and data[nx][ny] != 2:
        #사과가 없다면 이동하고 나서 원래 꼬리있던 자리 2 제거
            if data[nx][ny] == 0:
                data[nx][ny] = 2
                q.append((nx,ny))
                px, py = q.pop(0)
                data[px][py] = 0
        #사과가 있다면 사과는 먹고(머리자리 됨) 꼬리는 그대로 두기 
            if data[nx][ny] == 1:
                data[nx][ny] = 2
                q.append((nx,ny))
    #벽이나 몸통에 부딪혔을 때 
        else:
            time += 1
            break 
        x, y = nx, ny #다음위치로 머리 이동 
        time += 1 
        if index < 1 and time == info[index][0]: #회전해야하면 회전
            direction = turn(direction, info[index][1])
            index += 1
    return time 

print(sim())
```
